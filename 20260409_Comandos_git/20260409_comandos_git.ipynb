# Comando basicos do git

1. `git status`: mostra o status do repositório; exemplo: alteração arquivo no repositório.
2. `git diff`: mostra exatamente o que foi alterado no repositório; exemplo: alteraçao linha a linha nícel código. 
3. `git add`: adiciona alteracoes feitas em arquivos.
4. `git commit -m`: salva no historico do respositorio as alteracoes com uma mensagem dizendo o que foi feito; exemplo: 'alterei read.me'.
5. `git push`: se feito commits localmente envia essas alteracoes para o repositorio.
6. `git restore`: desfaz alteracoes feitas em um arquivo que ainda noa foram preparadas com git add, restaurando o arquivo para a ultima versao salva no commit; exemplo: se voce alterou main.py e se arrependendo antes de fazer git add, esse comando paga as mudancas.
7. `git restore --staged`: desfaz o git add, mas sem apgar as alteracoes feitas no arquivo; git restore --staged main.py tira main.py do proximo commit mas o conteudo alterado continua no arquivo.
8. `git pull`: baixa as atualizacoes do repositorio e integra as mudancas ao repositorio local.
9. `git log`: mostra o historico de commits do repositorio.
