## Exercicio 3

```python
def calcula(vetor1, vetor2):

    produto = np.dot(vetor1, vetor2)
    
    distancia = np.linalg.norm(vetor1 - vetor2)

    return produto, distancia
```

## Exercicio 4

```python
A = np.random.randint(-5,5,(4,4))

print("det:", np.linalg.det(A))

if np.linalg.det(A) != 0:
  print("É invertivel")

  A_inversa = np.linalg.inv(A)

  ident = np.dot(A, A_inversa)

  print("A * A_inversa:", ident)

else:
   print("A não é invertivel")
```

# Exercicio 5

```python
M = np.random.randn(5, 5)

B = (1/2) * (M + M.T)

autovalores, autovetores = np.linalg.eig(B)

print("Autovalores:" , autovalores)

print("Autovetores:", autovetores)

vet = autovetores[:, 0]
lam = autovalores[0]

print("Bv =", np.dot(B, vet))
print("lambda_v =", lam * vet)
```

# Matplot

```python
import matplotlib.pyplot as plt
```

## Exercicio 1

```python
df = pd.read_csv(r"C:\Users\Giovani\Downloads\iris.csv")
plt.hist(df['sepal.length'])
plt.title("Histograma Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequencia")
```

## Exercicio 2

```python
set = df[df['variety'] == 'Setosa']
versi = df[df['variety'] == 'Versicolor']
virg = df[df['variety'] == 'Virginica']

plt.scatter(set['sepal.length'], set['sepal.width'], label='Setosa')
plt.scatter(versi['sepal.length'], versi['sepal.width'], label='Versicolor')
plt.scatter(virg['sepal.length'], virg['sepal.width'], label='Virginica')

plt.title("Sepal Length versus Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
```

## Exercicio 3

```python
plt.scatter(df['petal.length'], df['petal.width'], s=df['sepal.length'])

plt.title("Petal Length versus Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
```

## Exercicio 4

```python
plt.subplot(1, 3, 1) 
plt.hist(set['petal.length'])
plt.title("Setosa")

plt.subplot(1, 3, 2)
plt.hist(versi['petal.length'])
plt.title("Versicolor")

plt.subplot(1, 3, 3)
plt.hist(virg['petal.length'])
plt.title("Virginica")
```

## Exercicio 5

```python

df_box_plot = [
    df['sepal.length'],
    df['sepal.width'],
    df['petal.length'],
    df['petal.width']
]

plt.boxplot(df_box_plot, labels=[
    'sepal.length',
    'sepal.width',
    'petal.length',
    'petal.width'
])

plt.title("Boxplot variedades Iris")
```

## Exercicio 6

```python
med = df.groupby('variety')[[
    'sepal.length',
    'sepal.width',
    'petal.length',
    'petal.width'
]].mean()

med.plot(kind='bar')

plt.title("Medias por variedade")
plt.xlabel("Variedade")
plt.ylabel("Valor medio")
```

## Exercicio 7

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
uniq = df['variety'].unique()

for varied in uniq:
    dados = df[df['variety'] == varied]
    ax.scatter(
        dados['sepal.length'],
        dados['sepal.width'],
        dados['petal.length'],
        label=varied
    )

ax.set_xlabel('sepal.length')
ax.set_ylabel('sepal.width')
ax.set_zlabel('petal.length')
ax.legend()
plt.show()
```
